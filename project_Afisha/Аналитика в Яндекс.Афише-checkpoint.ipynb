## Шаг 1. Загрузка данных и подготовка их к анализу

```python
# Загрузим библитеки
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import numpy as np
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# Читаем данные
visits = pd.read_csv(r'C:\Users\elizo\Desktop\Практикум\Проекты\.ipynb_checkpoints\Яндекс.Афиша\visits_log.csv')
orders = pd.read_csv(r'C:\Users\elizo\Desktop\Практикум\Проекты\.ipynb_checkpoints\Яндекс.Афиша\orders_log.csv')
costs = pd.read_csv(r'C:\Users\elizo\Desktop\Практикум\Проекты\.ipynb_checkpoints\Яндекс.Афиша\costs.csv')
```

```python
# Сделаем функцию для отображения данных и информации о фрейме
def show_info(df):
    """Функция для отображения первых пяти строк датафрейма и информации о датафрейме"""
    display(df.head(5))
    display(df.info())
```

```python
# Выведем информацию по каждому фрейму
for df in [visits,orders,costs]:
    show_info(df)
```

Таблица costs (информация о затратах на маркетинг):
- source_id — идентификатор рекламного источника
- dt — дата
- costs — затраты на этот рекламный источник в этот день

```python
# Переведем все названия столбцов к нижнему регистру.
visits.columns = visits.columns.str.lower().str.replace(' ', '_')
orders.columns = orders.columns.str.lower().str.replace(' ', '_')
display(visits.columns)
display(orders.columns)
```

```python
# Сделаем функцию, которая позволяет преобразовать колонку с датой в datetime
def convert_to_datetime_and_make_mth(df,column_name):
    """Функция для преобразования колонки column_name датафрейма df
    в дату и создания новой колонки month_column_name с месяцем на колонки column_name"""
    df = df.copy()
    df[column_name] = pd.to_datetime(df[column_name])
    return df
```

```python
# Применим функцию с помощью метода pipe()
visits = visits.pipe(convert_to_datetime_and_make_mth,'start_ts')
visits = visits.pipe(convert_to_datetime_and_make_mth,'end_ts')
orders = orders.pipe(convert_to_datetime_and_make_mth,'buy_ts')
costs = costs.pipe(convert_to_datetime_and_make_mth,'dt',)
```

**Вывод**

В первом шаге мы загрузил данные о визитах, заказах и расходах. Названия в колонках привел к нижнему регистру, в колонках с датами исправил тип с `object` на `datetime64`. В таблицах отсутствуют пропуски, можем приступать к анализу данных.

```python
# Чтобы посчитать недельное и месячное использование, выделим в отдельные столбцы год, месяц и неделю, а также полную дату.

visits['session_year']  = visits['start_ts'].astype('datetime64[Y]')
visits['session_month'] = visits['start_ts'].astype('datetime64[M]')
visits['session_week']  = visits['start_ts'].astype('datetime64[W]')
visits['session_date'] = visits['start_ts'].astype('datetime64[D]')

# Посчитаем метрики. Сгруппируем данные по уникальным пользователям и найдём среднее:
dau_total = visits.groupby('session_date').agg({'uid': 'nunique'}).mean()
wau_total = visits.groupby(['session_year', 'session_week']).agg({'uid': 'nunique'}).mean()
mau_total = visits.groupby(['session_year', 'session_month']).agg({'uid': 'nunique'}).mean()

# Сделаем срез по категории устройства touch:
dau_total_touch = visits.query('device == "touch"').groupby('session_date').agg({'uid': 'nunique'}).mean()
wau_total_touch = visits.query('device == "touch"').groupby(['session_year', 'session_week']).agg({'uid': 'nunique'}).mean()
mau_total_touch = visits.query('device == "touch"').groupby(['session_year', 'session_month']).agg({'uid': 'nunique'}).mean()

# Сделаем срез по категории устройства desktop:
dau_total_desktop = visits.query('device == "desktop"').groupby('session_date').agg({'uid': 'nunique'}).mean()
wau_total_desktop = visits.query('device == "desktop"').groupby(['session_year', 'session_week']).agg({'uid': 'nunique'}).mean()
mau_total_desktop = visits.query('device == "desktop"').groupby(['session_year', 'session_month']).agg({'uid': 'nunique'}).mean()

print('Среднее количество уникальных пользователей:')
print('В день:', int(dau_total))
print('В неделю:', int(wau_total))
print('В месяц:', int(mau_total))
print()
print('Среднее количество уникальных пользователей для Desktop:')
print('В день:', int(dau_total_desktop))
print('В неделю:', int(wau_total_desktop))
print('В месяц:', int(mau_total_desktop))
print()
print('Среднее количество уникальных пользователей для Touch:')
print('В день:', int(dau_total_touch))
print('В неделю:', int(wau_total_touch))
print('В месяц:', int(mau_total_touch))

# Параметры для отражения grid на графике
mpl.rcParams['axes.grid'] = True
mpl.rcParams['axes.grid.which'] = 'both'
mpl.rcParams['xtick.minor.visible'] = True

# Построим график количества посещений на сайт уникальных пользователей в месяц
visits.pivot_table(index = 'session_month', values = 'uid', aggfunc = 'count').\
plot(figsize = (15,6), grid=True, style='o-')
plt.title('Количество пользователей в месяц по устройствам')
plt.xlabel('Месяц')
plt.ylabel('Количество уникальных пользователей')
plt.show()

# Построим график количества посещений на сайт уникальных пользователей в месяц по категории устройства
visits.pivot_table(index = 'session_month', columns = 'device', values = 'uid', aggfunc = 'count').\
plot(figsize = (15,6), grid=True, style='o-')
plt.title('Количество пользователей в месяц по устройствам')
plt.xlabel('Месяц')
plt.ylabel('Количество уникальных пользователей')
plt.show()
```

**Сколько сессий в день?**

```python
# Сгруппируем посещения по устройствам и дате, посчитаем кол-во сессий уникальных пользователей в день

sessions_per_user = visits.groupby(['device','session_date']).agg({'uid': ['count','nunique']})
sessions_per_user.columns = ['n_sessions', 'n_users']

# Разделим число сессий на количество пользователей за день
sessions_per_user['sess_per_user'] = sessions_per_user['n_sessions'] / sessions_per_user['n_users']

# Сделаем срез по категории устройства и найдем среднее кол-во сессий в день
sessions_touch = sessions_per_user.query('device == "touch"').agg({'sess_per_user':'mean'})
sessions_desktop = sessions_per_user.query('device == "desktop"').agg({'sess_per_user':'mean'})

# Построим график кол-ва сессий в день по устройствам 
mpl.rcParams['axes.grid'] = False
sessions_per_user.pivot_table(index = 'session_date', columns = 'device', values = 'sess_per_user').\
plot(figsize = (15,5), ylim=(0,2))
plt.title('Среднее кол-во сессий на 1 пользователя в день по устройствам')
plt.xlabel('День')
plt.ylabel('Количество сессий')

print('Среднее количество сессий в день на 1 пользователя:', round(sessions_per_user['sess_per_user'].mean(),2))
print('Среднее количество сессий в день на 1 пользователя через устройства Touch:', sessions_touch[0].round(2))
print('Среднее количество сессий в день на 1 пользователя через устройства Desktoph:', sessions_desktop[0].round(2))
```

**Выводы**

**Как часто люди возвращаются?**

```python
# Возьмём дату, когда пользователь впервые проявил активность на сайте.
first_activity_date = visits.groupby(['uid'])['start_ts'].min()
first_activity_date.name = 'first_activity_date'

# Объединим first_activity_date с исходным датафреймом visits
visits = visits.join(first_activity_date,on='uid')

# Получим день начала месяца, за которую произошло событие 
visits['first_activity_month'] = visits['first_activity_date'].astype('datetime64[M]')

# Для каждой строки датафрейма можно рассчитать lifetime пользователя в рамках когорты
visits['cohort_lifetime'] = (visits['session_month'] - visits['first_activity_month']) / np.timedelta64(1,'M')
visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype('int')

# Сгруппируем данные по категории устройства, когорте и lifetime
# Посчитаем для каждой когорты количество активных пользователей на определённый «месяц жизни»:
cohorts = visits.groupby(['device', 'first_activity_month', 'cohort_lifetime']).agg({'uid' : 'nunique'}).reset_index()

# Найдём исходное количество пользователей в когорте. Возьмём их число на нулевой месяц:
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_activity_month', 'device', 'uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'})

# Объединим данные по когортам с исходным количеством пользователей в когорте:
cohorts = cohorts.merge(initial_users_count,on=['first_activity_month', 'device'])

# Разделим количество активных пользователей в каждую из недель на исходное число пользователей в когорте
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users']
```

```python
# Построим сводную таблицу для категории устройств desktop
retention_pivot_desktop = cohorts.query('device == "desktop"').\
    pivot_table(index='first_activity_month',columns='cohort_lifetime',values='retention',aggfunc='sum')

retention_pivot_desktop.index = retention_pivot_desktop.index.strftime("%m/%Y")

# Cоздадим тепловую карту
plt.figure(figsize=(13, 9))
sns.heatmap(retention_pivot_desktop, annot=True, fmt='.1%', linewidths=1, linecolor='gray', vmax = 0.1)
plt.title('Количество вернувшихся пользователей из когорты в последующие периоды для Desktop')
plt.xlabel('Возраст когорт, месяц')
plt.ylabel('Когорта')
plt.show()
```

```python
print(f'Для устройств Desktop среднее количество вернувшихся пользователей из когорты в первый месяц: {retention_pivot_desktop[1].mean(axis=0):.1%}')
```

```python
# Построим сводную таблицу для категории устройств touch
retention_pivot_touch = cohorts.query('device == "touch"').\
    pivot_table(index='first_activity_month',columns='cohort_lifetime',values='retention',aggfunc='sum')

retention_pivot_touch.index = retention_pivot_touch.index.strftime("%m/%Y")

# Cоздадим тепловую карту
plt.figure(figsize=(13, 9))
sns.heatmap(retention_pivot_touch, annot=True, fmt='.1%', linewidths=1, linecolor='gray', vmax = 0.1)
plt.title('Количество вернувшихся пользователей из когорты в последующие периоды для Touch')
plt.xlabel('Возраст когорт, месяц')
plt.ylabel('Когорта')
plt.show()
```

```python
print(f'Для устройств Touch среднее количество вернувшихся пользователей из когорты в первый месяц: {retention_pivot_touch[1].mean(axis=0):.1%}')
```

**Выводы**

Максимальное значение Retention Rate у кагорты 09/2019, характерно для всех категорий устройств.
Также видим, что RR в первый месяц для устройств Desktop немного выше, чем у Touch. Независимо от типа устройства RR по когортам убывает с течением времени.

### Продажи

**Когда люди начинают покупать?**

```python
# Найдем время первой покупки для каждого пользователя. Объединим по uid. 
first_order_df = orders.groupby(['uid']).agg(first_order_date=('buy_ts','min'))
start_shopping_df = pd.merge(first_order_df, visits, on='uid')

# Найдем разницу между временем первой покупки и временем первого заказа. Результат выразим в днях.
start_shopping_df['days_first_order'] = (start_shopping_df['first_order_date'] - start_shopping_df['first_activity_date'])\
    / np.timedelta64(1,'D')

# Гистограмма количества дней с момента первого визита до первой покупки
start_shopping_df['days_first_order'].hist(bins=30, label='Заказы')
plt.legend(loc='upper right')
plt.title('Количество дней с момента первого визита до первой покупки')
plt.xlabel('Дни')
plt.ylabel('Заказы')
plt.show()

# Гистограмма количества дней с момента первого визита до первой покупки в зависимости от устройства
start_shopping_df.query('device == "desktop"')['days_first_order'].hist(bins=30, label=('Desktop'))
start_shopping_df.query('device == "touch"')['days_first_order'].hist(bins=30, label=('Touch'))
plt.legend(loc='upper right')
plt.title('Количество дней с момента первого визита до первой покупки')
plt.xlabel('Дни')
plt.ylabel('Заказы')
plt.show()

print('В среднем пользователи совершают первую покупку через {} дня с момента первого посещения сайта.'.\
      format(int(start_shopping_df['days_first_order'].mean())))
print('В среднем пользователи desktop совершают первую покупку через {} дня с момента первого посещения сайта.'.\
      format(int(start_shopping_df.query('device == "desktop"')['days_first_order'].mean())))
print('В среднем пользователи touch совершают первую покупку через {} дня с момента первого посещения сайта.'.\
      format(int(start_shopping_df.query('device == "touch"')['days_first_order'].mean())))
```

**Выводы**

Чаще всего пользователи совершают первую покупку в первый день посещения сайта, но в среднем пользователи совершают первый заказ на 34 день. Категория устройства не влияет на время совершения первой покупки.

**Сколько раз покупают за период?**

```python
# Сформируем когорту по месяцу
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')

# Найдем месяц первой покупки пользователя
first_orders = orders.groupby('uid').agg({'order_month': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_month']

# Количество покупателей за каждый месяц
cohort_sizes = first_orders.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']

# Месяц первой покупки каждого пользователя добавим в таблицу с заказами
orders_new = pd.merge(orders,first_orders, on='uid')

# Подсчитаем количество заказов
cohorts = orders_new.groupby(['first_order_month','order_month']).agg({'buy_ts': 'count'}).reset_index()

report = pd.merge(cohort_sizes, cohorts, on='first_order_month')

# Найдем возраст когорты
report['age'] = (report['order_month'] - report['first_order_month']) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')

# Найдем количество покупок
report['rate'] = report['buy_ts'] / report['n_buyers']

# Сводная таблица
report_pivot = report.pivot_table(index='first_order_month',
                                columns='age',
                                values='rate',
                                aggfunc='mean').cumsum(axis=1)
report_pivot.index = report_pivot.index.strftime("%m/%Y")

# Построим тепловую карту
plt.figure(figsize=(13, 9))
sns.heatmap(report_pivot, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
plt.title('Количество покупок за период')
plt.xlabel('Месяц заказа')
plt.ylabel('Первый месяц заказа')
plt.show()

# Посчитаем количество покупок за пол года
purchases = round(report_pivot[5].mean(),2)
print(f'В среднем один покупатель совершает {purchases} покупок за полгода.')
```

**Выводы**

У когорты 06/2017 самое большое число заказав на одного пользователя, в целом постепенно у всех когорт увеличивается число заказов на одного пользователя.

**Какой средний чек?**

**Сколько денег приносят? (LTV)**

```python
# Выделим месяц из даты в таблице с заказами
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')
costs['month'] = costs['dt'].astype('datetime64[M]')

# Дата первой покупки каждого покупателя
first_orders = orders.groupby('uid').agg({'order_month': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_month']

# Посчитаем количество новых покупателей за каждый месяц
cohort_sizes = first_orders.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']

# Добавим месяц первой покупки каждого покупателя в таблицу с заказами
orders_new = pd.merge(orders,first_orders, on='uid')

# Сгруппируем таблицу заказов по месяцу первой покупки и месяцу каждого заказа и сложим выручку
cohorts = orders_new.groupby(['first_order_month','order_month']).agg({'revenue': 'sum'}).reset_index()

# Добавим в таблицу cohorts данные о том, сколько людей первый раз совершили покупку в каждый месяц
report = pd.merge(cohort_sizes, cohorts, on='first_order_month')

# LTV считают не из выручки, а из валовой прибыли. Значит, нужно найти валовую прибыль, умножив выручку на маржинальность
# Поскольку даных в проекте нет, считаем margin_rate = 100%
# LTV удобнее смотреть по «возрастным» когортам, где в столбцах — не месяц заказа, а возраст (age) когорты
report['age'] = (report['order_month'] - report['first_order_month']) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')

# Найдём LTV. Разделим валовую прибыль когорт за каждый месяц на общее число пользователей в каждой когорте
report['ltv'] = report['revenue'] / report['n_buyers']
    
# Посчитаем итоговый LTV когорт, сложим их LTV по месяцам   
output = report.pivot_table(
        index='first_order_month', 
        columns='age', 
        values='ltv', 
        aggfunc='mean').round(2).cumsum(axis=1)

 
output.index = output.index.strftime("%m/%Y")


plt.figure(figsize=(13, 9))
mpl.rcParams['axes.grid'] = False
sns.heatmap(output, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
plt.title('Итоговый LTV когорт по месяцам')
plt.xlabel('Возраст когорты, в месяцах')
plt.ylabel('Когорты')
plt.show()

print(f'В среднем пользователь приносит {round(output[5].mean(axis=0),2)} у.е. за 6 месяцев.')
```

```python
# Дата первой покупки каждого покупателя
first_orders = orders.groupby('uid').agg({'order_month': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_month']

# Определим источник для каждого посетителя, отсортируй визиты по start_ts, 
# сгруппируй по uid и примени к столбцу source_id функцию first()
first_source_df = visits.sort_values(by = 'start_ts', ascending = True).groupby(['uid']).agg({'source_id' : 'first'}) 
orders_source_df = pd.merge(left = orders, right = first_source_df, how = 'left', on = 'uid')

# Добавим колонку с источниками в дату первого заказа
first_orders_source_df = pd.merge(left = first_orders, right = first_source_df, how = 'left', on = 'uid')

# Посчитаем количество новых покупателей за каждый месяц
cohort_sizes_source = first_orders_source_df.groupby(['first_order_month', 'source_id']).agg({'uid': 'nunique'}).reset_index()
cohort_sizes_source.columns = ['first_order_month', 'source_id', 'n_buyers']

# Добавим месяц первой покупки каждого покупателя в таблицу с заказами
orders_source_df = pd.merge(orders_source_df,first_orders, on='uid')

# Сгруппируем таблицу заказов по месяцу первой покупки и месяцу каждого заказа и сложим выручку
cohorts_source_df = orders_source_df.groupby(['source_id', 'first_order_month','order_month']).agg({'revenue': 'sum'}).reset_index()

# Добавим в таблицу cohorts данные о том, сколько людей первый раз совершили покупку в каждый месяц
report_source_df = cohorts_source_df.merge(cohort_sizes_source, on=['first_order_month', 'source_id'])

# Найдём LTV. Разделим валовую прибыль когорт за каждый месяц на общее число пользователей в каждой когорте
report_source_df['ltv'] = report_source_df['revenue'] / report_source_df['n_buyers']

# Посчитаем итоговый LTV когорт, сложим их LTV по месяцам   
ltv_output_source = report_source_df.pivot_table(
        index='first_order_month', 
        columns='source_id', 
        values='ltv', 
        aggfunc='mean').cumsum(axis=0).round(2)

ltv_output_source.index = ltv_output_source.index.strftime("%m/%Y")

plt.figure(figsize=(13, 9))
mpl.rcParams['axes.grid'] = False
sns.heatmap(ltv_output_source, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
plt.title('Итоговый LTV когорт по источникам')
plt.xlabel('Источник')
plt.ylabel('Когорты')
plt.show()
```

**Сколько стоило привлечение одного покупателя из каждого источника?**

```python
# Выделим месяц из даты в таблице с заказами
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')
costs['month'] = costs['dt'].astype('datetime64[M]')

# Дата первой покупки каждого покупателя
first_orders = orders.groupby('uid').agg({'order_month': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_month']

# Посчитаем количество новых покупателей за каждый месяц
cohort_sizes = first_orders.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']

# Добавим месяц первой покупки каждого покупателя в таблицу с заказами
orders_new = pd.merge(orders,first_orders, on='uid')

# Сгруппируем таблицу заказов по месяцу первой покупки и месяцу каждого заказа и сложим выручку
cohorts = orders_new.groupby(['first_order_month','order_month']).agg({'revenue': 'sum'}).reset_index()

# Добавим в таблицу cohorts данные о том, сколько людей первый раз совершили покупку в каждый месяц
report = pd.merge(cohort_sizes, cohorts, on='first_order_month')

# LTV считают не из выручки, а из валовой прибыли. Значит, нужно найти валовую прибыль, умножив выручку на маржинальность
# Поскольку даных в проекте нет, считаем margin_rate = 100%
# LTV удобнее смотреть по «возрастным» когортам, где в столбцах — не месяц заказа, а возраст (age) когорты
report['age'] = (report['order_month'] - report['first_order_month']) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')

# Найдём LTV. Разделим валовую прибыль когорт за каждый месяц на общее число пользователей в каждой когорте
report['ltv'] = report['revenue'] / report['n_buyers']

# считаем расходы за месяц
monthly_costs = costs.groupby('month').sum()

# Добавим данные о расходах в отчёт и посчитаем CAC
report_new = pd.merge(report, monthly_costs, left_on='first_order_month', right_on='month')
report_new['cac'] = report_new['costs'] / report_new['n_buyers']


# Посчитаем итоговый CAC когорт, сложим их CAC по месяцам   
output_cac = report_new.pivot_table(
        index='first_order_month', 
        columns='age', 
        values='cac', 
        aggfunc='mean').round(2).cumsum(axis=1)

 
output_cac.index = output_cac.index.strftime("%m/%Y")


plt.figure(figsize=(13, 9))
mpl.rcParams['axes.grid'] = False
sns.heatmap(output_cac, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
plt.title('Итоговый CAC когорт по месяцам')
plt.xlabel('Возраст когорты, в месяцах')
plt.ylabel('Когорты')
plt.show()

print(f'Привлечение одного покупателя в среднем обошлось {round(output_cac[5].mean(axis=0),2)} у.е. за 6 месяцев.')
```

**На сколько окупились расходы? (ROI)**

```python
# Рассчитаем ROI: поделим LTV на CAC
report_new['roi'] = report_new['ltv'] / report_new['cac']

output_roi = report_new.pivot_table(
        index='first_order_month', 
        columns='age', 
        values='roi', 
        aggfunc='mean').cumsum(axis=1)
 
output_roi.index = output_roi.index.strftime("%m/%Y")

plt.figure(figsize=(13, 9))
sns.heatmap(output_roi, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
plt.title('ROI когорт по месяцам')
plt.xlabel('Возраст когорты, в месяцах')
plt.ylabel('Когорты')
plt.show()

print(f'В среднем у когорт к концу 6 месяца ROI: {round(output_roi[5].mean(axis=0),2)}')
```

```python
# Посчитаем ROI по источникам

# Объединим ранее рассчитанные по источникам данные заказов(ltv) с расходами(cac)
romi_source_df = report_source_df.merge(cac_source_df, left_on=['first_order_month','source_id', 'n_buyers'],\
                                                        right_on=['first_order_month','source_id', 'n_buyers'])

romi_source_df['age'] = (romi_source_df['order_month'] - romi_source_df['first_order_month']) / np.timedelta64(1, 'M')
romi_source_df['age'] = romi_source_df['age'].round().astype('int')

# Посчитаем метрики: LTV, СAC, ROI по источника
romi_source_df['ltv'] = romi_source_df['revenue'] / romi_source_df['n_buyers']
romi_source_df['cac'] = romi_source_df['costs'] / romi_source_df['n_buyers']
romi_source_df['roi'] = romi_source_df['ltv'] / romi_source_df['cac']

# Построим сводную таблицу по источникам
romi_source_pivot = romi_source_df.pivot_table(
        index='first_order_month', 
        columns='source_id', 
        values='roi', 
        aggfunc= 'sum').cumsum(axis=0)

romi_source_pivot.index = romi_source_pivot.index.strftime("%m/%Y")

plt.figure(figsize=(13, 9))
sns.set(style='white')
sns.heatmap(romi_source_pivot, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
plt.title('ROI по источникам')
plt.xlabel('Источник')
plt.ylabel('Месяц привличения клиента')
plt.show()
```
